### As we observed above, the color of the car has been changed in the image which was ready via cv2 but displayed via matplotlib. This is because CV2 reads the data as BGR and matplotlib reads as RGB. Which means the red and blue channel get interchanged in both the libraries.

```python
## Now let's convert the BGR image into RGB
image_BGR_TO_RGB=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
## Display the image
plt.imshow(image_BGR_TO_RGB)
```

```python
## Let's conver the image into RGBA
image_BGR_TO_RGBA=cv2.cvtColor(image, cv2.COLOR_BGR2RGBA)
plt.imshow(image_BGR_TO_RGBA)
```

### RGBA stands for RED, GREEN, BLUE, ALPHA. The alpha channel stands for the transparency. By changing the levels of it we can make image look transparent, translucent and opaque.

```python
## Let's conver the image into gray
image_BGR_TO_GRAY= cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(image_BGR_TO_GRAY)
```

```python
def visualize_RGB_channels(imageArray=None, figsize=(10,7)):
  ## Let's split the RGB channels
  B,G,R=cv2.split(imageArray)

  ## Create a zero matrix of any channel's size so that whenever we want we can only display red and green channels and replace blue channel with this zero matrix
  Z=np.zeros(B.shape, dtype=B.dtype)   

  fig, ax=plt.subplot((2,2), figsize=figsize)     ## 2X2 because we want 4 photos, original, red, green and blue 

  ## printing ax.shape
  [axi.set_axis_off() for axi in ax.ravel()]      ## for removing the x and y axis from the picture

  ax[0,0].set_title("original image")
  ax[0,0].imshow(cv2.merge((R,G,B)))

  ax[0,1].set_title("Red channel")
  ax[0,1].imshow(cv2.merge((R,Z,Z)))

  ax[1,0].set_title("Green channel")
  ax[1,0].imshow(cv2.merge((Z,G,Z)))

  ax[1,1].set_title("Blue channel")
  ax[1,1].imshow(cv2.merge((Z,Z,B)))

```

```python
visualize_RGB_channels(imageArray=image)
```

```python
def visulize_RGB_channels(imgArray=None, fig_size=(10,7)):
  # splitting the RGB components
  B, G, R = cv2.split(imgArray)

  # zero matrix
  Z = np.zeros(B.shape, dtype=B.dtype)

  # init subplots
  fig, ax = plt.subplots(2,2, figsize=fig_size)

  
  # print(ax.shape)
  [axi.set_axis_off() for axi in ax.ravel()]

  ax[0,0].set_title("original image")
  ax[0,0].imshow(cv2.merge((R,G,B)))

  ax[0,1].set_title("red image")
  ax[0,1].imshow(cv2.merge((R,Z,Z)))

  ax[1,0].set_title("green image")
  ax[1,0].imshow(cv2.merge((Z,G,Z)))

  ax[1,1].set_title("blue image")
  ax[1,1].imshow(cv2.merge((Z,Z,B)))
```

```python
visualize_RGB_channels(imageArray=image)
```
